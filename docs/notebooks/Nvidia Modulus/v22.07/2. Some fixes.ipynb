# Problems with Apptainer
By default the Apptainer binds certain volumes and imports all the environment variables from the host system. 

* A list of what is binded: https://apptainer.org/docs/user/main/bind_paths_and_mounts.html#system-defined-bind-paths
* The `/usr/local/etc/apptainer/apptainer.conf` allows the administrator to modify the default behavoiour.
* You can use `--no-mount` flag to manually disable the volume mounting and `--bind` to only bind specific volumes as follows:

`apptainer shell --no-mount hostfs,tmp,sys,dev,proc,sys,devpts,home,slave --bind "$(pwd)":/data modulus.img`

This automatic binding is complete madness. Recently, I modified the `$HOME/.bashrc` inside the Apptainer image and It broke my Host system's Python installation. So, It is better to stop all the automatic volume binding.

* One can disable all the automatic volume binds using `--contain` flag. The specific [documentation](https://apptainer.org/docs/user/main/docker_and_oci.html#default-mounts-home).
* One can also disable the environment variables import using `--cleanenv` or `-e`. The specific [documenation](https://apptainer.org/docs/user/main/docker_and_oci.html#environment-propagation).

If we don't disable the environment propagation, the image will always point to the Host computer's HOME directory. See the following output wher I compared an Apptainer image with a Docker image.

# Using the GPU
One can use the [GPU inside the Apptainer image](https://apptainer.org/docs/user/1.0/gpu.html#requirements) using `--nv` flag.

So the command will look like this:

```sh
apptainer shell --nv --contain --cleanenv --bind "$(pwd)":/data,/tmp:/tmp  modulus.img 
```
Here we can replace `shell` with `shell` or `exec`. A similar [documentation](https://docs.crc.nd.edu/resources/singularity.html#apptainer-on-gpu-nodes).

# Downloading Modulus examples
If we don't download Modulus example repository with Git LFS we will get [this error](https://forums.developer.nvidia.com/t/unable-to-run-helmholtz-py-after-installing-22-03-1/221919/5). So first install Git LFS from [this page](https://git-lfs.github.com/). After running the `install.sh` file, run `git lfs install` in the terminal. Then pull the Modulus exanple using `git clone`
